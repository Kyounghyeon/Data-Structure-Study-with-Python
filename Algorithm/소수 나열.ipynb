# 소수 나열1
- 1000 이하의 소수 나열
- 전체 비교

```python
count = 0
# 1000까지 탐색
for num in range(2,1001):
    comp = 0
    for i in range(2,num):
        
        """
        num을 num미만의 수로 나누었을 때 나누어 떨어지는 수가 있다면
        즉, 나머지가 0이라면 comp에 1을 더하기
        comp가 0이라면 나누어 떨어지는 수가 없는 것이므로 소수
        """
        if (num % i == 0):
            comp += 1 
        count+=1
        
    if (comp == 0):
        print(num)
print(f"총 비교 횟수: {count}")
        
```

# 소수 나열 2
- 1과 자기 자신나누어 떨어지는 수가 적어도 하나 존재한다면   
소수가 아니니까 하나라도 나온다면 비교 그만

```python
# 소수 나열 2

count = 0
# 1000까지 탐색
for num in range(2,1001):

    for i in range(2,num):
        
        """
        num을 num미만의 수로 나누었을 때 나누어 떨어지는 수가 하나라도 있다면,
        for loop 중단, num은 소수가 아님.
        """
        count += 1
        if (num % i == 0):
            break 
            
    else:
        print(num)
print(f"총 비교 횟수: {count}")
        
```

# 소수 나열3 
- 알고리즘을 개선해보자
- 짝수는 절대 안됨
- 홀수만 판단
- 3의 배수인가? 5의 배수인가? 7의 배수인가? 9는 3의 배수이므로 제외
- 11의 배수인가?
- ---> 다른 소수로 나누어 떨어지는가를 볼 것

```python
# 소수 나열 3 
prime=[2]
count = 0

# 짝수 제외, 홀수만 탐색
for num in range(3,1001, 2):
    for i in prime:
        count += 1
        # 소수로 나누었을 때 떨어지는 수는 소수가 아님
        if num % i == 0 :
            break
    else:
        prime.append(num)

print(prime)        
print(f"총 비교 횟수: {count}")
```

# 소수 나열 4

- 알고리즘을 개선해보자
- 짝수 안되고,
- 소수로만 나눠도 판단 가능
- 근데 예를 들어서 33의 경우 3x11인데, 3(소수)로 나눠도 11(소수)로 나눠도 나누어 떨어짐. 
  이건 불필요한 나눗셈. 즉 3으로만 나눠도 충분함
- sqrt(n)이하의 소수로만 나누자
- 직사각형의 넓이를 가로와 세로로 나타낸다고 생각하면 편함

```python
import math

prime = [2]
count = 0
for num in range(3, 1001, 2):
    
    for denom in prime:
        
        if denom <= math.floor(math.sqrt(num)):
            count += 1
            if num % denom == 0:
                
                break

    else:
        prime.append(num)
```
